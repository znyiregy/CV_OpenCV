## Cat face detection with  OpenCV


```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("10006.jpg")
```

```python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, "gray")
plt.show()
```

```python
classifier = cv2.CascadeClassifier("../data/haarcascades/haarcascade_frontalcatface_extended.xml")
```

```python
faces = classifier.detectMultiScale(gray, minNeighbors=1)
```

```python
print(faces)
```

```python
c = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(c, (x, y), (x + w, y + h), (0, 255, 0), 2)
    print(face)

i = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```
